```python
import pandas as pd
import eurostat
```

```python
data = eurostat.get_data('tin00174')
df = pd.DataFrame(data)

# Normanization of column names
df.columns = df.iloc[0]
df = df.iloc[1:]

ref = list(df)
def starts_20(w):
    return w.startswith("20")
res_2 = filter(starts_20, ref)
print(res_2)

for value in res_2: 
    df[value] = df[value].astype(float)
    
df = df[df["geo\TIME_PERIOD"].str.len()  < 3]
df = df.drop[]

print(df.dtypes)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(df)
```
